```python
import yfinance as yf
import pandas as pd
import numpy as np
import datetime
%matplotlib inline
```

```python
start = datetime.datetime(2006,1,1)
end = datetime.datetime(2016,1,1)
```

```python
BAC = yf.download('BAC',start,end)

# CitiGroup
C = yf.download('C',start,end)

# Goldman Sachs
GS = yf.download('GS',start,end)

# JPMorgan Chase
JPM = yf.download('JPM',start,end)

# Morgan Stanley
MS = yf.download('MS',start,end)

# Wells Fargo
WFC = yf.download('WFC',start,end)
```

```python
tickers = ['BAC','C','GS','JPM','MS','WFC']
```

```python
bank_stocks = pd.concat([BAC,C,GS,JPM,MS,WFC],axis=1,keys=['BAC','C','GS','JPM','MS','WFC'])
```

```python
bank_stocks
```

```python
bank_stocks.columns.names = ['Bank Ticker','Stock_info']
bank_stocks.head()
```

```python
bank_stocks.xs(key='Close',axis=1,level='Stock_info').max()
```

```python
returns = pd.DataFrame()
```

```python
for x in tickers:
    returns[x+' Return'] = bank_stocks.xs(key = x,axis=1,level='Bank Ticker')['Close'].pct_change()
returns.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
g=sns.PairGrid(returns[1:])
g.map_diag(sns.histplot,kde=False,bins=10)
g.map_upper(plt.scatter,s=5)
g.map_lower(plt.scatter,s=5)
```
